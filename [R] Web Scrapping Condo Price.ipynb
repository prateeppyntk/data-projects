# **Import Data of Living Insider's Website**

```r
url <- "https://www.livinginsider.com/living_project_en/5/303/Condo/Buysell/all/1/Condo-Ideo-Q-Chula-Samyan.html"
```

```r
price <- url %>%
  read_html() %>%
  html_elements("div.t-16") %>%
  html_text2()
View(price)
glimpse(price)
```

```r
type <- url %>%
  read_html() %>%
  html_elements("div.listing-tag") %>%
  html_text2()
View(type)
glimpse(type)
```

```r
details <- url %>%
  read_html() %>%
  html_elements("div.item-desc") %>%
  html_text2()
View(details)
glimpse(details)
```

# **Create Dataframe**

```r
df <- data.frame(price,type,details)
```

# **Clean Data #1**

```r
df <- df %>%
  separate(col = details, into = c("others","indeed"), sep = "Siam Paragon ,Chulalongkorn,Samyan *")
```

```r
df <- df %>%
  separate(col = indeed, into = c("size","indeed2"), sep = " Sq.m. *")
```

```r
df$indeed2 <- df$indeed2 %>%
  str_replace_all(.,"Level"," ")
```

```r
df$indeed2 <- df$indeed2 %>%
  str_replace_all(.," 1 Beds"," ")
```

```r
df$indeed2 <- df$indeed2 %>%
  str_replace_all(.," 1 Bath"," ")
```

```r
df$indeed2 <- df$indeed2 %>%
  str_replace_all(.," Studio room"," ")
```

```r
df$indeed2 <- df$indeed2 %>%
  str_replace_all(.," 2 Beds"," ")
```

```r
df$indeed2 <- df$indeed2 %>%
  str_replace_all(.," 2 Bath"," ")
```

```r
df <- df %>%
  separate(col = others, into = c("c1","v1"), sep = "[0-9]{4}-[0-9]{2}-[0-9]{2} [0-9][0-9]:[0-9][0-9]")
```

```r
df <- df %>%
  separate(col = c1, into = c("c2","v2"), sep = "yesterday *")
View(df)
```

```r
df <- df %>%
  separate(col = c2, into = c("c3","v3"), sep = "[0-9]+ hours *")
View(df)
```

```r
df %>%
  select(price, type, v1, v2, v3, size, indeed2)
```

# **Combine Data of 3 Columns (from v1, v2, v3 into view) & Create running number**

---



```r
df <- df %>%
  select(price, type, v1, v2, v3, size, indeed2) %>%
  mutate(new_v = coalesce(v1,v2))
```

```r
df <- df %>%
  select(price, type, v1, v2, v3, size, indeed2, new_v) %>%
  mutate(views = coalesce(new_v,v3))
```

```r
number <- 1:48
```

```r
df <- data.frame(number,df)
View(df)
```

# **Clean Data #2**

```r
df$views <- df$views %>%
  str_replace_all(.,"Foreigner Quota"," ")
```

```r
df$size <- df$size %>%
  as.numeric()
```

```r
df$price <- df$price %>%
  parse_number() %>%
  as.numeric()
```

```r
c <- df %>%
  select(number,views) %>%
  filter(grepl("k",views))
```

```r
c$views <- c$views%>%
  parse_number() %>%
  as.numeric()
```

```r
c <- c %>%
  select(number,views) %>%
  mutate(new_v = views * 1000)
cn <- View(c$number)
cv <- View(c$new_v)
```

```r
df <- df%>%
  select(number,
      selling_price = price,
      selling_type = type,
      room_size_sqm = size,
      floor = indeed2,
      old_v = views)
  View(df)
```

```r
df2 <- df %>%
  left_join(c, by="number")
```

```r
df2[cn,"old_v"] <- cv
View(df2)
```

```r
df2$old_v <- df2$old_v %>%
  as.numeric()
View(df2)
```

```r
ideo_CU <- df2 %>%
  select(number,
    selling_price,
    selling_type,
    room_size_sqm,
    floor,
    view_qty = old_v) %>%
  arrange(selling_type, -view_qty)
```
