```python
# import pandas
import pandas as pd
import random
```

```python
# create function of pao ying chub game
def pao_ying_chub() :

    while True :
        start = input('\nAre you ready?\n(Yes: yes / No: no)\n\n')

        # create list of choice for players
        bot = ['Rock', 'Paper', 'Scissors']

        # create variables for received score
        player_scores = 0
        bot_scores = 0

        if start == 'yes':
            print('\nWelcome to Pao Ying Chub Game!!\nThese are choices for selecting : Rock, Paper, Scissors\n')

            while True :
                player_ans = input('\nPlease input your answer\n(If you want to stop this game, Input exit for stop.)\n=>')\
                .strip()

                # create summary table when user type exit
                if player_ans == 'exit' :
                    sums = pd.DataFrame({'player':player_scores,'bot':bot_scores},index=['scores'])
                    if player_scores > bot_scores :
                        print("\nCongratuation!!! You Are Win.\n")
                        return sums
                    else :
                        print("\nYou Are Loss, Please play with us again.\n")
                        return sums

                # condition game between input and random answer of bot
                else :
                    bot_ans = random.sample(bot,k=1)
                    print(f'Answer of Bot is {bot_ans}.')
                    if player_ans == 'Rock' :
                        if bot_ans == ['Scissors'] :
                            player_scores +=1
                            print("You are Win!")
                            print(f"Your current score is {player_scores} and Bot's score is {bot_scores}.")
                        elif bot_ans == ['Paper'] :
                            bot_scores +=1
                            print("You are Loss!")
                            print(f"Your current score is {player_scores} and Bot's score is {bot_scores}.")
                        else :
                            print("You fare with Bot.")
                            print(f"Your current score is {player_scores} and Bot's score is {bot_scores}.")
                    elif player_ans == 'Paper' :
                        if bot_ans == ['Rock'] :
                            player_scores +=1
                            print("You are Win!")
                            print(f"Your current score is {player_scores} and Bot's score is {bot_scores}.")
                        elif bot_ans == ['Scissors'] :
                            bot_scores +=1
                            print("You are Loss!")
                            print(f"Your current score is {player_scores} and Bot's score is {bot_scores}.")
                        else :
                            print("You fare with Bot.")
                            print(f"Your current score is {player_scores} and Bot's score is {bot_scores}.")
                    elif player_ans == 'Scissors' :
                        if bot_ans == ['Paper'] :
                            player_scores +=1
                            print("You are Win!")
                            print(f"Your current score is {player_scores} and Bot's score is {bot_scores}.")
                        elif bot_ans == ['Rock'] :
                            bot_scores +=1
                            print("You are Loss!")
                            print(f"Your current score is {player_scores} and Bot's score is {bot_scores}.")
                        else :
                            print("You fare with Bot.")
                            print(f"Your current score is {player_scores} and Bot's score is {bot_scores}.")
                    else :
                        print('Please type again.')
        elif start == 'no':
            return('See you next times.')
        else :
            print('Please type again.')
```
