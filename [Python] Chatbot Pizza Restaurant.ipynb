```python
# import pandas
import pandas as pd
```

```python
from numba.core.ir import Return
# create order_pizza() function
def order_pizza() :

    # dictionary of pizza's menu
    pizza = {'id' : [1,2,3,4,5],
             'name' : ['Hawaiian','Super Deluxe','Seafood Deluxe','Seafood Cocktail','Shrimp Cocktail'],
             'price' : [100,120,150,150,200]
            }

    # create blank list for keep menu details
    l_menu = []

    # create menu table and convert to dataframe
    for o in [0,1,2,3,4] :
       l_menu.append((pizza["id"][o], pizza["name"][o], pizza["price"][o]))
    l_menu = pd.DataFrame(l_menu,columns=['menu_id','menu_name','menu_price']).reset_index(drop=True)

    # run into function
    while True :
        start = input('Welcome! Do you need to order with us now?\n(Yes: yes / No: no)\n\n')

        # condition from input for starting
        if start == 'yes' :
            print('This is our pizza menu:')
            print(l_menu)

            # crate variables for received value from function
            ids = []
            orders = []
            qtys = []
            prices = []
            amount = []
            sums = { 'id' : ids,
                     'order_name' : orders,
                     'order_qty' : qtys,
                     'unit_price' : prices,
                     'total_amount' : amount }

            while True :
                order = input('\nWhat would you like to order?\n(Yes: type menu_id / No: type exit)\n\n')

                # condition from input for ordering
                if order == 'exit':

                 # create condition for separate customer who has other order or not have
                 if orders == [] :
                  return "\nPlease come back again next time.\n"

                 else :
                    discount = input('\nDo you have any discount?\n(Yes: yes / No: type any key)\n\n')

                    # condition from input for check discount
                    if discount == 'yes' :

                        while True :
                            code = input('\nPlease input your code?\n(Enter = pizza50 / Enter = nocode for without discount)\n\n')

                            # condition from input for received discount code
                            if code == 'pizza50' :
                                s = pd.DataFrame(sums)
                                s = s.groupby(['id','order_name','unit_price'])['order_qty','total_amount']\
                                .sum().sort_values('id').reset_index()
                                s['total_amount'] = s['total_amount']*0.95
                                return s
                            elif code == 'nocode' :
                                s = pd.DataFrame(sums)
                                return s.groupby(['id','order_name','unit_price'])['order_qty','total_amount']\
                                .sum().sort_values('id').reset_index()
                            else :
                                print('Please type again.')
                    else :
                        s = pd.DataFrame(sums)
                        return s.groupby(['id','order_name','unit_price'])['order_qty','total_amount']\
                        .sum().sort_values('id').reset_index()

                elif order in ['1','2','3','4','5'] :
                    ids.append(pizza['id'][(int(order))-1])
                    orders.append(pizza['name'][(int(order))-1])
                    prices.append(pizza['price'][(int(order))-1])
                    qty = input('\nHow many quantity?\nPlease order 0-9 pcs/time\n\n')

                    if int(qty) in range(0,9) :
                        qtys.append(int(qty))
                        amount.append(pizza["price"][(int(order))-1]*(int(qty)))
                        check_bill = input('\nDo you need to check bill? or continue to order?\n(Yes: check / Do Not: type any key)\n\n')
                        if  check_bill == 'check' :
                            discount = input('\nDo you have any discount?\n(Yes: yes / No: type any key)\n\n')

                            if discount == 'yes' :

                                while True :
                                    code = input('\nPlease input your code?\n(Enter = pizza50 / Enter = nocode for without discount)\n\n')

                                    if code == 'pizza50' :
                                        s = pd.DataFrame(sums)
                                        s = s.groupby(['id','order_name','unit_price'])['order_qty','total_amount']\
                                        .sum().sort_values('id').reset_index()
                                        s['total_amount'] = s['total_amount']*0.95
                                        return s
                                    elif code == 'nocode' :
                                        s = pd.DataFrame(sums)
                                        return s.groupby(['id','order_name','unit_price'])['order_qty','total_amount']\
                                        .sum().sort_values('id').reset_index()
                                    else :
                                        print('Please type again.')
                            else :
                                s = pd.DataFrame(sums)
                                return s.groupby(['id','order_name','unit_price'])['order_qty','total_amount']\
                                .sum().sort_values('id').reset_index()

                    else :
                        print('\nPlease type again.\n')

                else :
                    print('\nPlease type again.\n')

        elif start == 'no' :
            return "\nPlease come back again next time.\n"
        else :
            print('\nPlease type again.\n')
```
